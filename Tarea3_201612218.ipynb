# Eliminación de datos duplicados e incompletos

- dropna, elimina los datos incompletos 
- drop_duplicate, este elimina datos duplicados.

```python

datos = datos.dropna() 

datos = datos.drop_duplicates()
```

# Graficos 

- Barras

```python
sns.barplot(x="citric acid", y="residual sugar", data=datos)
plt.title("Gráfico de Barras")
plt.xlabel("Acidez citrica")
plt.ylabel("Azucar residual")

plt.show()
```

- Disperción

```python
plt.figure(figsize=(6,4))

sns.scatterplot(x="pH", y="alcohol", data=datos, hue="quality", palette="Set2", s=100)
plt.title("Relación entre pH y Alcohol")
plt.xlabel("pH")
plt.ylabel("Alcohol (%)")
plt.show()
```

- BlockSpot



```python
sns.boxplot(x="quality", y="alcohol", data=datos)
plt.title("Distribución de Alcohol por Calidad del Vino")
plt.xlabel("Calidad")
plt.ylabel("Alcohol (%)")
plt.show()
```

# Conclusiones.

- La matriz de correlación ver que algunas variables como alcohol y quality están ligeramente correlacionadas, mientras que otras no tienen relación clara.

- En el Gráfico de barras se comparan promedios o totales por categorías.

- Al combinar estos gráficos y análisis, se obtiene una visión completa del dataset
